## Euclidean Distance
What people think of when they think of distance. Distance from point x to y, given by

$$d(\overrightarrow{x},\overrightarrow{y}) = \sqrt{(x_1-y_1)^2+(x_2-y_2)^2}$$

However this case is only for two dimensions. We can generalize this formula to the following

$$ d(\overrightarrow{x}, \overrightarrow{y}) = \left[\sum_{i=1}^n\left|x_i-y_i\right|^2\right]^{1/2}$$

where $$\overrightarrow{x} = <x_1, x_2, ... , x_n>,\\
\overrightarrow{y} = <y_1, y_2, ... , y_n>,\\
\text{and }\forall i(x_i, y_i \in \Re)$$

## Lp Norms 
Lp norms are special cases of euclidian distances that takes everything to the P power instead. 

$$ d(\overrightarrow{x}, \overrightarrow{y}) = \left[\sum_{i=1}^n\left|x_i-y_i\right|^p\right]^{1/p}$$

If p = 1, we have Manhattan Distance<br>
if p = 2, we have euclidian distance <br>
if p = 1 and $x_i, y_i \in \{0,1\}$ we have Hamming distance


## Weighting dimensions
Sometimes dimensions/attributes are not evenly distributed. For example the read dot belongs in the left circle but is closer to the right one. We can fix this by weighting the dimensions of the circles. 


This slightly modifies our previous equation:

$$ d(\overrightarrow{x}, \overrightarrow{y}) = \left[\sum_{i=1}^nw_i\left|x_i-y_i\right|^p\right]^{1/p}$$

And we get ellipses as seen above.

### Mahalanobis distance
Now we have a concept. a circle is same euclidean distance from the center, but we can find a Mahalanobis distance such that A and B are equidistant 


The calculation of this number involves a variable $S$ called the covariance matrix and is calculated from the data distribution

$$ d(\overrightarrow{x},\overrightarrow{y}) = \sqrt{(\overrightarrow{x}-\overrightarrow{y})^TS^{-1}(\overrightarrow{x}-\overrightarrow{y})}$$

## Metrics

We have a few qualities that qualify somethign as a metric, otherwise it is a measure

$$ d(x,y) = 0 \text{ iff } x = y \text{    reflextivity} \\
d(x,y) \geq 0 \text{    non-negative}\\
d(x,y) = d(y,x) \text{    symmetry}\\
d(x,y) + d(y,z) \geq d(x,z) \text{    triangle inequality}$$

## Missing Data
What if data is not given for some category?

There are a few things we can do
* Discard all examples missing the category
* Fill in the blanks with mean value
* only use a category in the distance measure if both examples give a value

$$ w_i =
  \begin{cases}
    1       & \quad \text{if both } x_i \text{ and } y_i \text{ are defined}\\
    0  & \quad \text{else}\\
  \end{cases}
$$

$$d(\overrightarrow{x},\overrightarrow{y}) = \frac{n}{n-\sum_{i=1}^n w_i} \left[\sum_{i=1}^n w_i\phi(x_i,y_i)\right]$$
