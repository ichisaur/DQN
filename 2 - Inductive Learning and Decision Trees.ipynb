# Inductive Learning and Decision Trees

# Inductive Learning
Tasks have the inputs and outputs clearly stated

    Like predicting stock markets bast on past price data
        You input a date and ticker symbol
        The Output is if the price is higher or lower or what the price change will be
    Or predicting outcomes of sporting events
        Input is the game, opponents, date
        Ouput is which team will win
However, you have to define the the things in tasks. The following are not tasks

    Studying the relationship between weather and sports game outcomes
    Applying neural networks to natural language processing
        No clearly defined inputs and outputs
***
Instances are like a single input. Like weather on specific days. 


Functions take those inputs and gives out an output. Like if aperson decides to go out to lunch with me on a specific day given the weather


Inductive learning tries to predict a new output for a new instance. 


## How this works:
Define a set X of instances. Each instance is a tuple with the n parameters. Like days descriped by forcast attributes.

A function then takes X and outputs Y

    Goes to lunch -> 0 or 1
    Response to Lunch -> Set of possible Responses
        ie. {"No", "Yes", "How about tomorrow"}
    Probability of lunch -> [0,1]

We give it a set of training examples D

    D is exampels of the target function

It finds a hypothesis h such that h(x) is approximate to f(x)

It is important that we note that we are restricting ourselves to the Hypothesis space H of the function f.

Exampels of this type of inductive learning are  

    Credit Risk Analysis:
        X -> properties of customer and purchase        
        f(x) -> Approve or Disapprove        
    Face Recognition:
        X -> Bitmap of Face
        f(x) -> Who that person is
    Automatic Steering
        X ->Bitmap of road surface
        f(x) -> How much to turn the wheel

It is appropriate to use inductive learning to build a facial recognizer, but not for a calculator. In that case you would just write a calculator program

The question becomes: What general characteristics make aproblem suitable for inductive learning?

Things where

    There is no human expert
    Humans can do the tasks but cant describe the rpocess

    each user needs a custimized function
