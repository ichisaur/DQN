# Regression Learning

Given a set of possible examples $$X = \{\textbf{x}_\textbf{1} \textbf{,...,x}_\textbf{n} \}$$ 

and given that each example is a vector of attributes $$\textbf{x}_i = <x_{i1} ,..., x_{ik}>$$

Basically this is a set containing sets of attributes.

Our target function f maps X to Y where X is the input and we want to get Y values as the output

$$f:X \to Y$$

The data we input into the sy stem are tuples of x and  y `<`x,y`>` for each example in X

We need to find a hypothesis such that for all x, h(x) is approximately equal to f(x)

Basically generate a function h that approximates f close enough to what we desire. 

$$ \forall \textbf{x},h(\textbf{x}) \approx f(\textbf{x})$$


## Assumption
We assume that the response is a linear fucntion of the input plus some gaussian noise

This means that we can approximate the actual funciton using a  linear combination of inputs

Observed response $$y = f(\textbf{x})+\epsilon$$ where $$\epsilon \sim \mathcal{N}(0,\sigma ^2)$$

## Hypothesis Space

Basically each hypothesis has a weight w attached to it, and our job is to find an appropriately good w that minimizes error

$$h(\textbf{x}) = w_0 + w_1x_1+w_1x_2+...w_kx_k$$
$$\textbf{w} = <w_0, w_1,...,w_k>$$

There si actually a solution to this  problem that minimizes the sum of the residuals

$$\textbf{w} = (\textbf{X}^T\textbf{X})^{-1}\textbf{X}^T\textbf{y}$$


## Polynomial Regression

We can use the exact same concepts to create a polynomial regression line instead of a linear one. 

$$h(x) = w_0 + w_1z+w_2z^2 +...+w_kz^k$$
$$\textbf{w} = (\textbf{X}^T\textbf{X})^{-1}\textbf{X}^T\textbf{y}$$


Going up to higher K values for the function, essentially creating kth order polynomials will reduce the error, however, this will cause overfitting 


but this all depends on how many data points there are. The more data points, the higher your k value can go.

As we increase the complexity K, the bias decreases, but the variance icnreases.
