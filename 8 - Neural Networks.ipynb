# Learning Through Examples

## Perceptrons
First wave of neural network algorithms

Basically, let f be a function from $X = <x_1, x_2,...> \text{where } x\in\{0,1\} \text{to } y\in\{0,1\}$

Given training data $\{(X_1,y_1),(X_2,y_2)...\} \text{We learn } h(X) \text{, an approximation of } f(X)$

We introduce a bias $x_0$ that always has a value of 1. This bias and the rest of the inputs have weights attatched tho them $<w_0,w_1,w_2...>$ such that if the following expression is true

$$\sigma =
\begin{cases}
    1       & \quad \text{if } \sum_{i=0}^nw_ix_i > 0\\
    0  & \quad \text{else}\\
  \end{cases}
  $$
  

From this simpleexample we can solve for some logic operators. However, the problem iwht perceptrons are that they can only represent linearly seperable fuctions.

Of course, a combination of perceptrons can solve almost anything.

### Solving for the weights
We initialize $w_0,w_1=0$
Then we cycle through the learning examples with the following in mind

$$w_i \leftarrow w_i+\Delta w_i$$
$$\Delta w_i = \eta (t-o)x_i$$
where
* $\eta$ is the learning Rate
* $t$ is the target value
* $o$ is the perceptron output
* $x_i$ is the input value

### Whats bad about them?
We can always plug in multiple perceptrons to make any function, but deciding what the weights are are difficult. <br>
Assignments of error to parental inputs becomes a problem. 

Perceptrons also use a step function, meaning a small change in input either makes no change in output or a huge change. 

Although we have multiple ways we can help solve this situation.For example we can use a linear or differentiable function to represent the problem instead of a step one. 

Varying the input creats a perceptible change in output and we can characterize how the error changes 

## Gradient Descent for Linear Functions
The output function, using a linear function, becomes 
$$\sigma(\overrightarrow{x}) = \overrightarrow{w}\cdot\overrightarrow{x}$$

And we measure Error using the function
$$E(\overrightarrow{w}) \equiv \frac{1}{2} \sum_{d\in D}(t_d-o_d)^2$$

where
* $d \in D$ is data
* $t_d$ is target value
* $o_d$ is linear unit output

## Pros and Cons of each metho
Perceptron Rule and Threshold Units
* Converges on answer only if data is linearly seperable
* cant assign error to parent nodes

Gradient Descent
* Localy minimizes error, even if not lienarly seperable
* works for multi layer netoworks, but linear units only make linear decision surfaces. Meaning XOR is still impossible 
* Step function sin't differentiable

# Sigmoid Logistic Unit

$$output = \sigma (net) = \frac{1}{1+e^{-net}}$$
where
$$net = \sum_{i=0}^{n}w_ix_i$$

It has a differentiable function, allowing for gradient descent, and can be used to learn non linear functions.


Multiple layers can be combined together to get an a correct output through hidden layering.



Notice how the function could get stuck in local minima

Finding the gradientThis is possible because finding the differential of a Sigmoid is easy
$$\frac{\partial\sigma(y)}{\partial y} = \sigma(y)\cdot(1-\sigma(y))$$

There's a backpropogation algorithm, but we like literally never use it  because we can just use a package to do it for us. 

We should also normalize inputs to lie in [0,1] range

Problems that arrise with Neurla Netorks are how slow they are to train, and local minima screwing up the error surface used to find the minimum error.

However, given enough hidden functions, we can approximate any continuous function. We don't use millions of units because they are slow and efficieniency would be out of whack. Overfitting would also be a problem then. 

## Combatting overfitting
Two methods
* Early Slopping, like using a lot of hidden units. But we have to begin to not overuse this. 
* Cross validation to choose the right amount of hidden units.

## Learning Rate

Learning rate is a knob basically that you can twist. Look for the validation set accuracy to decrease and stabalize then halve the learning rate is a nice option.

In modern neural networks, local minima is not much of an issue. Early stopping is useful. Other times, initailizing at zero is what you want, and you want just keep training until you are at zero error. Multiple layers are common, and those layers may be combined by convolution, or max pooling
