# Instance Based Learning

## Basic Principle
You have a set of examples that are already solved. ie they already have known outputs

Whenever you recieve a new input, you can look at the examples that already exist to assign an output that closely resembles the most similar known case
***
Example: Disease diagnoses
You have a set of known diagnoses based of symptomology
You see a patient with symptoms that most closely resemble case X
Case X was the flu
Therefore you have the flue

## How this works in a mathematical sense

You have a set of sample data $$X = \{\overrightarrow{x}_i\}$$

Each example is a n-tuple of attributes
$$\overrightarrow{x}_1 = <a_1,...,a_k>$$
There is a function that maps X onto Y
$$f:X\to Y$$
and the Data we have is a set of duples of `<`example, target function value`>`
$$ D = \{<\overrightarrow{x}_1,f(\overrightarrow{x}_1)>,...,<\overrightarrow{x}_m,f(\overrightarrow{x}_m)>\}$$
We find a hypothesis h such that
$$\forall \overrightarrow{x},h(\overrightarrow{x}) \approx f(\overrightarrow{x})$$


## How do we actually do this?
It is actually pretty simple. We just look for the nearest neighbor in k-space. Given all the data points, we scan through the list finding the nearest point and assign that point's value to our current point

A good example of this would be voronoi diagrams


## Distinguishing Characteristics
### Instance based learner

    we consider typically the Euclidean distance measure to find the nearest neighbor. 
    We only consider one. 
    We don't use a weighting function. 
    And the output is the same as the nerest neighbor. 
    
### K nearest neighbor

Exactly the same as Instance based, only we consider the nearest K neighbors <br> 
We don't weight, but we vote using the K nearest neighbors or take average

## Pros and Cons

* Advantages
** Fast to train, lazy method
** Learns compled functions easily
** Doesn't lose information

* Disadvantages
** Slow at qurerying stuff
** Lots of storage space
** Can be fooled by irrelevant redundant attributes

Also know that lower level intutions of geometry doesn't extend well to higher levels, so it is easy to be misled

## A better method: Kernel Weighted Regression
* Use sclaled euclidean to measure distance
* Consider all of the neightbors
* We have an weighting function that states that nearby points are weighted strongly, far points are weighted weakly (Kw is kernel width)
$$ w_i = exp\left({\frac{-d(x_i, x_q)^2}{K_W^2}}\right)$$
* And we fit with the neighbors using a weighted average

$$h(x_q) = \frac{\sum_i w_i \bullet f(x_i)}{\sum_i w_i}$$
