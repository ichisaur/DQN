# What is it?
Collaborative filtering is a system that filters things based on your past history

There are two approaches to this
* User based
** You will like something because other users like you liked that thing
* Item Based
** You will like something because you like things similar to that thing

A pretty big example of this is Amazon's recommendation system 

## Feature Selection
We have to decide on which features to measure and which attributes to discard when measuring the similarities of users or items

We also have to consider if there are features that are relatively unimportant to the task at hand. 

Use user ratings and purchase history to measure user preference. 

## How do we measure similarity?

We can use a distance measurement in N-dimensional space



We can also use similarity measurments like p-norm, pearson correlation, cosine similarity

## Things to consider

We have to deal with missing values, not having any prior history, which similarity measure to use, how many neighbors to pick, how to weigh them, and if we should use user based or item based algorithms to make our decision.
